```python
import sys
import os
import glob
import h5py
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import rfft, irfft
from matplotlib.animation import FuncAnimation

current_dir = os.getcwd()
sys.path.append(os.path.abspath(os.path.join(current_dir, '..')))

eta_hat, phi_hat, Hs, Tp, modes, time, length, x = None, None, None, None, None, None, None, None

with h5py.File("Z:\\files\\simulation.h5", "r") as data:
    eta_hat = data["eta_hat"][:]
    phi_hat = data["phi_hat"][:]
    Hs = data["Hs"][:]
    Tp = data["Tp"][:]
    time = data["time"][:]

    modes = data.attrs["modes"]
    length = data.attrs["length"]
    Ta = data.attrs["Ta"]
    x = np.linspace(0, length, 2*modes)

index = np.argmin(np.abs(time - 2*Ta))

eta_hat = eta_hat[:, index:, :]
phi_hat = phi_hat[:, index:, :]
time = time[index:] - time[index]
```

```python
import torch
import torch.nn as nn
from torchdiffeq import odeint
from FNO import FNO1d

class ODEFNOFunc(nn.Module):
    def __init__(self, modes=16, width=64, depth=4, size=1024):
        super().__init__()
        self.fno = FNO1d(in_channels=1, out_channels=1, width=width, modes=modes, depth=depth)
        self.size = size  # Input length

    def forward(self, t, y):
        # y: (B, N)
        return self.fno(y)  # Also returns (B, N)

end_index = 20
t = torch.from_numpy(time[:end_index]).float()
y0 = torch.from_numpy(irfft(eta_hat[0, 0, :])).float().unsqueeze(0)
y_true = torch.from_numpy(irfft(eta_hat[0, :end_index, :])).float().unsqueeze(1)

func      = ODEFNOFunc(modes=512, width=64, depth=4, size=1024)
optimizer = torch.optim.Adam(func.parameters(), lr=1e-3)
criterion = nn.MSELoss()

best_loss = np.inf
early_stop_count = 0
early_stopping_rounds = 25

for epoch in range(5000):
    optimizer.zero_grad()
    y_pred = odeint(func, y0, t)

    loss = criterion(y_pred, y_true)
    loss.backward()
    optimizer.step()

    if loss.item() < best_loss:
        early_stop_count = 0
        best_loss = loss.item()
        torch.save(func.state_dict(), "best.pt")
        print(f"Saved new best model at epoch {epoch} with loss {best_loss:.6f}")
    else:
        early_stop_count += 1
        if early_stop_count > early_stopping_rounds:
            print("Early Stopping!")
            break

    print(f"Epoch {epoch:4d} loss = {loss.item():.6f}")
    if epoch % 10 == 0:
      plt.plot(y_pred.detach().numpy()[end_index-1, 0, :])
      plt.plot(y_true.detach().numpy()[end_index-1, 0, :])
      plt.show()
```

```python
func.load_state_dict(torch.load("best.pt"))
t = torch.from_numpy(time[:]).float()
y_pred = odeint(func, y0, t).detach().numpy()[:, :, :]
y = irfft(eta_hat[0, :, :])
```

```python
import matplotlib.animation as animation

fig, ax = plt.subplots(figsize=(10, 4))
line1, = ax.plot([], [], label="Ground Truth")
line2, = ax.plot([], [], label="Prediction")
ax.set_xlim(x[0], x[-1])
ax.set_ylim(np.min(y), np.max(y))
ax.grid()
ax.legend()

def update(frame):
    line1.set_data(x, y[frame, :])
    line2.set_data(x, y_pred[frame, :])
    ax.set_title(f"Time: {time[frame]:.1f}s")
    return line1, line2

ani = animation.FuncAnimation(
    fig, update, frames=y.shape[0],
    blit=True, interval=50
)

ani.save("prediction_vs_truth_v2.mp4", fps=20, dpi=200)
plt.show()
```
